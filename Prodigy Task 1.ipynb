```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#taken an example of student data set
file_path = 'student-por.csv'
df = pd.read_csv(file_path)
df
```

```python
#barchart on sex of student
gender_counts = df['sex'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color=['magenta', 'blue'])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Distribution of Gender in the Dataset')
plt.show()
```

```python
#histogram on age group
age_group = df[(df['age'] >= 15) & (df['age'] <= 22)]['age']
plt.hist(age_group, bins=8, color='blue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age')
plt.show()
```

```python
#whether student lives in urban or rural
address_counts = df['address'].value_counts()
plt.bar(address_counts.index, address_counts.values, color=['blue', 'red'])
plt.xlabel('Address')
plt.ylabel('Count')
plt.title('Distribution of Address Values')
plt.show()
```

```python
#histogram on traveltime of student
plt.hist(df['traveltime'], bins=4, color='blue', edgecolor='black')
plt.xlabel('Travel Time')
plt.ylabel('Count')
plt.title('Histogram of Travel Time')
plt.show()
```

```python
#histogram on studytime of student
plt.hist(df['studytime'], bins=4, color='blue', edgecolor='black')
plt.xlabel('Study Time')
plt.ylabel('Count')
plt.title('Histogram of Study Time')
plt.show()
```

```python
#failures of student
plt.hist(df['failures'], bins=4, color='blue', edgecolor='black')
plt.xlabel('Failure')
plt.ylabel('Count')
plt.title('Histogram of Student failures')
plt.show()
```

```python
#first period grade
plt.hist(df['G1'], bins=4, color='blue', edgecolor='black')
plt.xlabel('First Period Grade')
plt.ylabel('Count')
plt.title('First Period Grade of Students')
plt.show()
```

```python
#second period grade
plt.hist(df['G2'], bins=4, color='blue', edgecolor='black')
plt.xlabel('Second Period Grade')
plt.ylabel('Count')
plt.title('Second Period Grade of Students')
plt.show()
```

```python
#last period grade
plt.hist(df['G3'], bins=4, color='blue', edgecolor='black')
plt.xlabel('Last Period Grade')
plt.ylabel('Count')
plt.title('Last Period Grade of Students')
plt.show()
```
